```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('student_data.csv')
df.head()
```

```python
sns.set()
_ = plt.hist(df['Target'])
_ = plt.xlabel('Target')
_ = plt.ylabel('Frequency')

plt.show()
```

```python
sns.set()
sns.swarmplot(x='Target', data=df)
plt.xlabel('Target')
plt.ylabel('Frequency')

plt.show()
```

```python
x = np.sort(df['Target'])
y = np.arange(1, len(x)+1) / len(x)

_ = plt.plot(x,y, marker='.', linestyle='none')
_ = plt.xlabel('Target')
_ = plt.ylabel('Frequency')
plt.margins(0.02)
plt.show()
```
